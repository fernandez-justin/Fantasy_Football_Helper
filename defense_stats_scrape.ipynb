```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def build_url(year,week):
    '''
    builds the url in order to get data
    '''
    url = 'https://fftoday.com/stats/playerstats.php?Season={}&GameWeek={}&PosID=99&LeagueID='.format(year,week)
    #url = 'https://fftoday.com/stats/playerstats.php?Season=2019&GameWeek=2&PosID=99&LeagueID='
    return url

def get_text(url):
    '''
    gets the text from the website in order
    to get the table of defensive statistics
    '''
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    tag_and_text = soup.find_all(class_='sort1')
    return tag_and_text
    
def get_df_from_text(year,week):
    '''
    creates df from the information grabbed from the site
    '''
    url = build_url(year,week)
    tag_and_text = get_text(url)
    master_list = []
    sub_list = []
    counter = 0
    # iterate through the soup that we grabbed and 
    # create a list containing all data
    for line in tag_and_text:
        counter += 1
        text = line.getText()
        if text[-1].isalpha():
            sub_list.append(text.split(' ')[1])
        else:
            sub_list.append(text)
        if counter == 13:
            master_list.append(sub_list)
            sub_list = []
            counter = 0
    # 13 data points per team
    column_names = ['team','g','sack','fr','int','deftd','pa','payd/g','ruyd/g','safety','kicktd','fpts','fpts/g']
    df = pd.DataFrame(master_list,columns = column_names)
    df['week'] = week
    df['year'] = year
    return df
```

```python
start_year = 2008
end_year = 2019
start_week = 1
end_week = 16
master_df = pd.DataFrame()
```

```python
for year in range(start_year,end_year+1):
    for week in range(start_week,end_week+1):
        df = get_df_from_text(year,week)
        print(year,week)
        #print(df)
        #print(master_df.head(1))
        master_df = master_df.append(df)
```

```python
master_df.to_csv('def_stats.csv')
```
